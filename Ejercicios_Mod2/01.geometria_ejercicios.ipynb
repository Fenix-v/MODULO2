    - $P1(-2, -2)$

```python
P1 = (-2, -2)
P2 = (2, 5)

# Punto 1
plt.plot(P1[0], P1[1], marker = "o", color = "blue") 

# Punto 2
plt.plot(P2[0], P2[1], marker = "o", color = "red") 

# Recta
plt.plot([P1[0], P2[0]], [P1[1], P2[1]], color = "green")

# Eje X
plt.axhline(0, color = "black")

# Eje Y
plt.axvline(0, color = "black")

plt.grid()
plt.show()
```

### Ejercicio 2:
- Usando la ecuación de la recta, encuentra la recta de los dos puntos del ejercicio anterior.

$$
y = \frac{(y_2 - y_1)(x - x_1)}{x_2 - x_1} + y_1
$$

- Usando _**x = np.linspace(-10, 10, 2)**_ grafica la recta.
- Grafica también los dos puntos.

```python
P1 = (-2, -2) # tupla
P2 = (2, 5) # tupla
x = np.linspace(-10, 10, 2) # genera valores de x en un rango deseado
y = ( ( (P2[1] - P1[1]) * (x - P1[0]) ) / ( P2[0] - P1[0] ) ) + P1 [1]
print(f"x: {x}")
print(f"y: {y}")

# Punto 1
plt.plot(P1[0], P1[1], marker = "o", color = "blue") 

# Punto 2
plt.plot(P2[0], P2[1], marker = "o", color = "red") 

# Recta
plt.plot(x, y)

# Eje X e Y
plt.axhline(0, color = "black")
plt.axvline(0, color = "black")

# grid o cuadrícula
plt.grid()
plt.show()
```

### Ejercicio 3:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de números enteros aleatorios entre el 1 y el 10.000 usando la librería _**random**_.

```python
import random

def generar_numeros_aleatorios(numero):
    lista = []
    for i in range(numero):
        numero_aleatorio = random.randint(1, 10_000)
        lista.append(numero_aleatorio)
        
    return lista
```

```python
generar_numeros_aleatorios(5)
```

```python
def generar_numeros_aleatorios(numero):
    return [random.randint(1, 10_000) for x in range(numero)]
```

```python
generar_numeros_aleatorios(8)
```

### Ejercicio 4:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de 0's y 1's de forma aleatoria usando la librería _**random**_.

```python
def generar_numeros_0_1(n):
    lista = []
    opciones = [0, 1]
    for i in range(n):
        lista.append(random.choice(opciones))
    return lista

generar_numeros_0_1(8)
```

```python
def generar_numeros_0_1(n):
    return [random.choice([0, 1]) for x in range(n)] # list comprehension: se hace cuando en el bucle hay poco código

generar_numeros_0_1(8)
```

### Ejercicio 5:

- Define una función que reciba dos listas y que calcule la _**distancia Euclideana**_.

$$D(P1, P2) = \sqrt[]{\sum_{i = 0}^{n} (x_{i} - y_{i})^{2}}$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcular_dist_euclidiana(lista1, lista2):
    distancia = 0
    for i in range(len(lista1)):
        distancia += (lista1[i] - lista2[i]) ** 2
        
    return np.sqrt(distancia)
```

```python
def calcular_dist_euclidiana(lista1, lista2):
    distancia = 0
    for x, y in zip(lista1, lista2):
        distancia += (x - y)**2
        
    return np.sqrt(distancia)
```

```python
def calcular_dist_euclidiana(lista1, lista2):
    return np.sqrt(np.sum( [(x - y)**2 for x, y in zip(lista1, lista2)] ))
```

```python
lista1 = generar_numeros_aleatorios(10)
lista2 = generar_numeros_aleatorios(10)
print(lista1)
print(lista2)
print(calcular_dist_euclidiana(lista1, lista2))
```

### Ejercicio 4:

- Define una función que reciba dos listas y que calcule la _**distancia Manhattan**_.

$$\text {Manhattan} = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcula_distancia_manhattan(lista1, lista2):
 
    if len(lista1) != len(lista2):
        raise ValueError("Las listas deben tener la misma longitud")
    return sum(abs(a - b) for a, b in zip(lista1,lista2))
```

```python
punto1 = [1, 2, 3]
punto2 = [4, 0, -1]

distancia = calcula_distancia_manhattan(punto1, punto2)

print(f"La distancia Manhattan entre dos puntos es : {distancia}")
```
