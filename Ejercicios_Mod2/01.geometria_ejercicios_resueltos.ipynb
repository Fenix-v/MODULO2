$$
y = \frac{(y_2 - y_1)(x - x_1)}{x_2 - x_1} + y_1
$$

### Ejercicio 3:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de números enteros aleatorios entre el 1 y el 10.000 usando la librería _**random**_.

```python
def lista_aleatoria(n):
    
    lista = [random.randint(1, 10_000) for i in range(n)]
    
    return lista
```

```python
lista_aleatoria(10)
```

### Ejercicio 4:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de 0's y 1's de forma aleatoria usando la librería _**random**_.

```python
def lista_0_1(n):
    
    lista = [random.choice(seq = [0, 1]) for i in range(n)]
    
    return lista
```

```python
lista_0_1(10)
```

### Ejercicio 5:

- Define una función que reciba dos listas y que calcule la _**distancia Euclideana**_.

$$D(P1, P2) = \sqrt[]{\sum_{i = 0}^{n} (x_{i} - y_{i})^{2}}$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def dist_euclideana(X, Y):
    
    total = 0
    
    for x, y in zip(X, Y):
        
        total = total + (x - y)**2
        
    total = np.sqrt(total)
    
    return total

# def dist_euclideana2(X, Y):
    
#     return np.sqrt(np.sum([(x - y)**2 for x, y in zip(X, Y)]))
```

```python
lista1 = lista_0_1(10)
lista2 = lista_0_1(10)

dist_euclideana(lista1, lista2)
```

### Ejercicio 4:

- Define una función que reciba dos listas y que calcule la _**distancia Manhattan**_.

$$\text {Manhattan} = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def dist_manhattan(X, Y):
    
    total = 0
    
    for x, y in zip(X, Y):
        
        total = total + np.abs(x - y)
        
    return total

def dist_manhattan2(X, Y):
    
    return np.sum([np.abs(x - y) for x, y in zip(X, Y)])
```

```python
lista1 = lista_aleatoria(10)
lista2 = lista_aleatoria(10)

dist_manhattan(lista1, lista2)
```

```python
lista1 = lista_0_1(10)
lista2 = lista_0_1(10)

dist_manhattan(lista1, lista2)
```

### Ejercicio 5:

- Define una función que reciba dos listas y que calcule la _**distancia Chebyshev**_.

$$\text{Chebysev} = max(\left| x_{1} - x_{0} \right|, \left| y_{1} - y_{0} \right|, ..., \left| k_{1} - k_{0} \right|)$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def dist_chebyshev(X, Y):
    
    valor_max = 0
    
    for x, y in zip(X, Y):
        
        valor = np.abs(x - y)
        
        if valor > valor_max:
            
            valor_max = valor
            
    return valor_max

def dist_chebyshev2(X, Y):
    
    return np.max([np.abs(x - y) for x, y in zip(X, Y)])
```

```python
lista1 = lista_aleatoria(10)
lista2 = lista_aleatoria(10)

dist_chebyshev(lista1, lista2)
```

```python
lista1 = lista_0_1(10)
lista2 = lista_0_1(10)

dist_chebyshev(lista1, lista2)
```

### Ejercicio 6:

- Define una función que reciba dos listas y un valor $p$ y que calcule la _**distancia Minkowski**_.

$$
\text{Minkowski} = \left(\sum_{i=1}^{n}\left|x_i - y_i\right|^{p}\right)^{\frac{1}{p}}
$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def dist_minkowski(X, Y, p):
    
    total = 0
    
    for x, y in zip(X, Y):
        
        total = total + np.abs(x - y)**p
        
    total = total**(1/p)
    
    return total

def dist_minkowski2(X, Y, p):
    
    return np.power(np.sum([np.power(np.abs(x - y), p) for x, y in zip(X, Y)]), 1/p)
```

```python
lista1 = lista_aleatoria(10)
lista2 = lista_aleatoria(10)

dist_minkowski(lista1, lista2, 2)
```

```python
lista1 = lista_0_1(10)
lista2 = lista_0_1(10)

dist_minkowski(lista1, lista2, 2)
```
